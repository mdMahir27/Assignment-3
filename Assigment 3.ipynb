```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import r2_score
import numpy as np
```

```python
data = pd.read_csv(r"Iris.csv")
print(data.head())
```

```python
null_counts = data.isnull().sum()
print("Number of null values:\n", null_counts)
```

```python
#Q3: Dropping null values
data = data.dropna()
print(data)
```

```python
#Q4: Label encoding string values
label_columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm","PetalWidthCm", "Species"]
label_encoder = LabelEncoder()

for col in label_columns:
    data[col] = label_encoder.fit_transform(data[col])
print(data[col])
```

```python
#Q5: Converting string values to numbers
numeric_columns = data.select_dtypes(include=['number']).columns
data_numeric = data[numeric_columns]
print(data_numeric)
```

```python
#Q6: Applying MinMax scaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data_numeric)
print(data_scaled)
```

```python
features = [ 'SepalLengthCm' , 'SepalWidthCm' , 'PetalLengthCm', 'PetalWidthCm']

target= 'Species'

X = data[features]
y = data[target]

```

```python
linreg = LinearRegression()
linreg.fit(X,y)
y_pred = linreg.predict(X)
print (linreg.intercept_)
print (linreg.coef_)
print('\n')
print('Sepal.length = 0.47*Petal.Length + 0.59*Sepal.Width + 2.25')
print('\n')
print('Holding all other independent variable fixed, a 1 centimeter increase in petallength lead to a increase in Sepalength by 0.47 centimeter')
print('Holding all other independent variable fixed, a 1 centimeter increase in sepalwidth lead to a increase in Sepalength by 0.59 centimeter')
```

```python
#Q8: r2 score
r2_original = r2_score(y, y_pred)
print("R2 value without cross-validation:", r2_original)
```

```python
cv_scores=cross_val_score(linreg,X,y,cv=6,scoring='r2')
print(cv_scores)
```
